```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib as mat
import matplotlib.pyplot as plt
%matplotlib inline
```

## Abrindo o dataset

```python
#Abrindo o dataset
file_name='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/coursera/project/kc_house_data_NaN.csv'
df=pd.read_csv(file_name)
df.head(20)
```

* id :- It is the unique numeric number assigned to each house being sold.
* date :- It is the date on which the house was sold out.
* price:- It is the price of house which we have to predict so this is our target variable and aprat from it are our features.
* bedrooms :- It determines number of bedrooms in a house.
* bathrooms :- It determines number of bathrooms in a bedroom of a house.
* sqft_living :- It is the measurement variable which determines the measurement of house in square foot.
* sqft_lot : It is also the measurement variable which determines square foot of the lot.
* floors: It determines total floors means levels of house.
* waterfront : This feature determines whether a house has a view to waterfront 0 means no 1 means yes.
* view : This feature determines whether a house has been viewed or not 0 means no 1 means yes.
* condition : It determines the overall condition of a house on a scale of 1 to 5.
* grade : It determines the overall grade given to the housing unit, based on King County grading system on a scale of 1 to 11
* sqft_above : It determines square footage of house apart from basement.
* sqft_basement : It determines square footage of the basement of the house.
* yr_built : It detrmines the date of building of the house.
* yr_renovated : It detrmines year of renovation of house.
* zipcode : It determines the zipcode of the location of the house.
* lat : It determines the latitude of the location of the house.
* long : It determines the longitude of the location of the house.
* sqft_living15 : Living room area in 2015(implies-- some renovations)
* sqft_lot15 : lotSize area in 2015(implies-- some renovations)

## Limpando os dados

* Identificando os valores nulos do dataset, mas eles não serão tratados agora, e sim mais adiante

```python
#Identificando valores NaN
df.isnull().sum()
```

* A coluna Unnamed: 0 pode ter um nome mais apropriado e ser usada como um índice. É uma coluna desnecessária para o dataset

```python
#Transformando a coluna Unnamed:0 para House
df = df.rename(columns={'Unnamed: 0': 'House'})
df.head()
```

```python
#Definindo a coluna 'House' como índice
df.set_index('House', inplace=True)
df.head()
```

* É bom também identificar e retirar valores dupilcados, caso haja.

```python
#Retirando valores duplicados
df.drop_duplicates().head()
```

* Algumas colunas, como date, view e zipcode podem ser retirada, pois para este problema, não é obrigatório saber quando a casa foi construída ou quando ela apareceu no banco de dados, ou se ela já foi vista ou não. Isso deixará o dataset mais leve

```python
#Retirando as colunas desnecessárias
del df['date']
del df['view']
del df['zipcode']
df.head()
```

* Olhando o dataset, aparecem alguns dados não faz sentido, como por exemplo, nas colunas  bathrooms (banheiros) e floors (andares).
* Na coluna bathrooms, aparecem casas com 1.75, 2.25, 2.50 ou 4.50 banheiros, o que não faz sentido.
* Na coluna floors, aparecem casas com 1.5 andares, e também não faz sentido.
* Para resolver isto, esses atributos foram transformados de floats para inteiros, pois esses dados fracionados seriam arredondados para um número inteiro

```python
#Transformar as variáveis bathrooms e floors em valores inteiros
df['bedrooms'] = df['bedrooms'].values.astype(np.int64)
df['bathrooms'] = df['bathrooms'].values.astype(np.int64)
df['floors'] = df['floors'].values.astype(np.int64)
df.dtypes
```

* Agora sim os valores nulos podem ser retirados

```python
#Retirando todos os valores NaN
df.fillna(value='bedrooms', inplace=True)
df.isnull().sum()
```

```python
#Dados transformados
df.head(20)
```

* Por fim, serão identificados e eliminados os outliers, finalizando a primeira parte do projeto.

```python
#Valor médio das casas por período
a = df.loc[df['yr_built']<1910].price.mean()
b = df.loc[df['yr_built']>1911].price.mean() and df.loc[df['yr_built']<=1930].price.mean()
c = df.loc[df['yr_built']>1931].price.mean() and df.loc[df['yr_built']<=1950].price.mean()
d = df.loc[df['yr_built']>1951].price.mean() and df.loc[df['yr_built']<=1970].price.mean()
e = df.loc[df['yr_built']>1971].price.mean() and df.loc[df['yr_built']<=1990].price.mean()
f = df.loc[df['yr_built']>1991].price.mean() and df.loc[df['yr_built']<=2010].price.mean()
g = df.loc[df['yr_built']>2011].price.mean()

print('Preço médio das casas contruídas antes de 1910: ${:.2f}'.format(a))
print('Preço médio das casas contruídas entre 1911 e 1930: ${:.2f}'.format(b))
print('Preço médio das casas contruídas entre 1931 e 1950: ${:.2f}'.format(c))
print('Preço médio das casas contruídas entre 1951 e 1970: ${:.2f}'.format(d))
print('Preço médio das casas contruídas entre 1971 e 1990: ${:.2f}'.format(e))
print('Preço médio das casas contruídas entre 1991 e 2010: ${:.2f}'.format(f))
print('Preço médio das casas contruídas depois de 2010: ${:.2f}'.format(g))
```

```python
#Representação gráfica
ano = ['1900-1910', '1911-1930', '1931-1950', '1951-1970', '1971-1990', '1991-2010', '2010-atual']
valor = [a,b,c,d,e,f,g]
    
plt.figure(figsize=(10,10))
plt.bar(ano, valor)
plt.xticks(ano)
plt.ylabel('Valor médio')
plt.xlabel('Anos')
plt.title('Gráfico do preço médio por período')
plt.grid(True, lw = 0.5)
plt.show()
print()
```
