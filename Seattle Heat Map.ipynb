# Heat Map

```python
import pandas as pd
```

```python
df=pd.read_csv('kc_house.csv')
df.head()
```

```python
#Mapa de calor com a localização das casas
colunasSelecionadas = ['lat','long']
mapa_calor = df.filter(items=colunasSelecionadas)
mapa_calor.head()
```

```python
import os
import folium

from folium.plugins import HeatMap
m = folium.Map([47.6062, -122.3321], tiles='stamentoner', zoom_start=12)
HeatMap(mapa_calor).add_to(m)
m
```

```python
m.save('mapa-calor.html')
```
